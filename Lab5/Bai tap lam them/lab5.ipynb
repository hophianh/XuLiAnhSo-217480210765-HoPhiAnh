1 Chọn đối tượng trong ảnh

```python
import numpy as np
import imageio.v2 as iio
import matplotlib.pylab as plt

data = iio.imread('fruit.jpg')

# Mở rộng thêm vùng bên phải để lấy trọn quả táo xanh
green_apple = data[880:1080, 1590:1900]

iio.imsave('green_apple.jpg', green_apple)

plt.imshow(green_apple)
plt.show()
```

2 Tịnh tiến đơn 

```python
import numpy as np
import scipy.ndimage as nd
import imageio.v2 as iio
import matplotlib.pylab as plt
data = iio.imread('fruit.jpg', mode = 'F')
bdata = nd.shift(data, (50, -30))
plt.imshow(bdata)
plt.show()
```

3 Thay đổi kích thước ảnh

```python
import numpy as np
import scipy.ndimage as nd
import imageio.v2 as iio
import matplotlib.pylab as plt

data = iio.imread('fruit.jpg')

# Phóng to gấp 3 lần
phongto = nd.zoom(data, (3, 3, 1))
plt.imshow(phongto)
plt.title("Phóng to gấp 3 lần")
plt.axis('off')
plt.show()

# Thu nhỏ còn 0.3 lần
thunho = nd.zoom(data, (0.3, 0.3, 1))
plt.imshow(thunho)
plt.title("Thu nhỏ còn 0.3 lần")
plt.axis('off')
plt.show()
```

4 Xoay ảnh


```python
import numpy as np
import scipy.ndimage as nd
import imageio.v2 as iio
import matplotlib.pylab as plt

data = iio.imread('fruit.jpg')

# reshape=True
rotated1 = nd.rotate(data, 45, reshape=True)
plt.imshow(rotated1)
plt.title("Xoay 45 độ - reshape=True")
plt.show()

# reshape=False
rotated2 = nd.rotate(data, 45, reshape=False)
plt.imshow(rotated2)
plt.title("Xoay 45 độ - reshape=False")
plt.show()
```

5 Dilation và Erosion

```python
import numpy as np
import scipy.ndimage as nd
import imageio.v2 as iio
import matplotlib.pylab as plt

data = iio.imread('world_cup.jpg', mode='F')

# Dilation với size=(5,5)
structure = np.ones((5,5))
dilated = nd.binary_dilation(data, structure=structure)
plt.imshow(dilated)
plt.title("Dilation size=(5,5)")
plt.show()

# Erosion với size=(5,5)
eroded = nd.binary_erosion(data, structure=structure)
plt.imshow(eroded)
plt.title("Erosion size=(5,5)")
plt.show()
```
